```python
import numpy as np
from sklearn.neighbors import KernelDensity

# Sample data
df = np.array([10, 20, 30, 40, 50, 60]).reshape(-1, 1)

# Create KDE model
kde = KernelDensity(
    kernel='epanechnikov',
    bandwidth=10
)

# Fit model
kde.fit(df)

# Create points where density will be calculated
x = np.linspace(0, 70, 1000).reshape(-1, 1)

# Calculate log density
log_density = kde.score_samples(x)

# Convert log density to density
density = np.exp(log_density)

# Plot
plt.plot(x, density)
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Epanechnikov KDE")
plt.show()
```

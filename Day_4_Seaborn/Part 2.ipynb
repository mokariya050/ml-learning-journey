```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# catplot
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
```

```python
tips.head()
```

```python
# axes level function
sns.stripplot(data=tips,x='day',y='total_bill',palette=['red','black'],jitter=0.1,hue='sex')
plt.legend(loc='upper left')
plt.show()
```

```python
# figure level function
sns.catplot(data=tips,x='day',y='total_bill',palette=['red','black','green','blue'],kind='strip')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',palette=['green','black','red','yellow'],kind='swarm')
plt.show()
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',palette=['blue','red'],hue='sex')
plt.legend(loc='upper left')
plt.show()
```

```python
sns.boxplot(data=tips,x='sex',y='total_bill',palette=['red','green'])
plt.show()
```

```python
sns.catplot(data=tips,x='total_bill',kind='box',height=4,aspect=2)
plt.show()
```

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='box',height=4,aspect=2,row='day')
plt.show()
```

```python
sns.violinplot(data=tips,x='sex',y='total_bill')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='violin',hue='sex')
plt.show()
```

```python
import numpy as np
sns.barplot(data=tips,x='sex',y='total_bill',errorbar=('ci',0),width=0.5,hue='day',estimator=np.size)
plt.show()
```

```python
sns.barplot(data=tips,y='total_bill',x='sex')
```

```python
sns.pointplot(data=tips,y='total_bill',x='sex')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
plt.show()
```

```python
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
# hue parameter is available
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.residplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
# multi plots
g=sns.FacetGrid(data=tips,col='day',row='time',hue='smoker')
g.map(sns.boxplot,'sex','total_bill')
g.add_legend()
plt.show()
```

```python
# pair plot
iris
```

```python
sns.pairplot(data=iris,hue='species')
plt.show()
```

```python
g=sns.PairGrid(data=iris)
g.map(sns.scatterplot)
plt.show()
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

plt.show()
```

```python
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_length','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)

plt.show()
```

```python
sns.jointplot(data=iris,x='sepal_length',y='petal_length',kind='hex')
plt.show()
```

```python
g=sns.JointGrid(data=iris,x='sepal_length',y='petal_length',hue='species')
g.plot(sns.scatterplot,sns.boxplot)
plt.show()
```

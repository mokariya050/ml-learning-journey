```python
import numpy as np
import time
import sys
```

```python
# python list and numpy n d array 
a=[i for i in range(1000000)]
b=[i for i in range(1000000,2000000)]

start=time.time()
c=[]
for i in range(len(a)):
    c.append(a[i]+b[i])

time.time()-start
```

```python
a1=np.arange(1000000)
b1=np.arange(1000000,2000000)

start=time.time()
c1=a1+b1
time.time()-start
```

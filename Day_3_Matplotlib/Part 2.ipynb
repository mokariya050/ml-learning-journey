```python
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(12,10))

ax[0,0].scatter(data['avg'],data['strike_rate'])
ax[0,0].set_title("scatter plot")
ax[0,0].set_xlabel("avg")
ax[0,0].set_ylabel('strike_rate')

ax[0,1].hist(data['avg'])
ax[0,1].set_title("histogram")
ax[0,1].set_xlabel('avg')

ax[1,1].scatter(data['avg'],data['runs'])
ax[1,1].set_title("scatter plot")
ax[1,1].set_xlabel("avg")
ax[1,1].set_ylabel('runs')

ax[1,0].hist(data['strike_rate'])
ax[1,0].set_title("histogram")
ax[1,0].set_xlabel('strike_rate')

plt.show()
```

```python
fig=plt.figure(figsize=(12,10))
ax1=fig.add_subplot(2,2,1)

ax1.scatter(data['avg'],data['strike_rate'])
ax1.set_title("scatter plot")
ax1.set_xlabel("avg")
ax1.set_ylabel('strike_rate')

ax2=fig.add_subplot(2,2,2)
ax2.hist(data['avg'])
ax2.set_title("histogram")
ax2.set_xlabel('avg')

ax3=fig.add_subplot(2,2,3)
ax3.scatter(data['avg'],data['runs'])
ax3.set_title("scatter plot")
ax3.set_xlabel("avg")
ax3.set_ylabel('runs')

ax4=fig.add_subplot(2,2,4)
ax4.hist(data['strike_rate'])
ax4.set_title("histogram")
ax4.set_xlabel('strike_rate')

plt.show()
```

```python
fig=plt.figure(figsize=(10,11),constrained_layout=True)
ax1=fig.add_subplot(111,projection='3d')
ax1.scatter3D(data['runs'],data['avg'],data['strike_rate'],color='red')
ax1.set_title("scatter plot in 3D")
ax1.set_xlabel("runs")
ax1.set_ylabel('avg')
ax1.set_zlabel('strike_rate')
ax1.grid(False)
plt.show()
```

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx,yy=np.meshgrid(x,y)
```

```python
zz= xx**2  +  yy**2
zz.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111,projection="3d")
ax.plot_surface(xx,yy,zz,cmap='viridis')
plt.show()
```

```python
zz= np.sin(xx)  +  np.cos(yy)
zz.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111,projection="3d")
ax.plot_surface(xx,yy,zz,cmap='summer')
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111)
c1=ax.contour(xx,yy,zz,cmap='summer')
fig.colorbar(c1)
plt.show()
```

```python
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(111)
c1=ax.contourf(xx,yy,zz,cmap='viridis',levels=10)
fig.colorbar(c1)
plt.show()
```

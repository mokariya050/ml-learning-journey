```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"E:\Datasets\sharma-kohli.csv")
data.head()
```

```python
plt.style.use('classic')
plt.figure(figsize=(10,5))
plt.plot(data['index'],data['RG Sharma'],label="rohit",color="red",linestyle="-",marker="o",markersize=5,markerfacecolor="yellow")
plt.plot(data['index'],data['V Kohli'],label="virat",color="green",linestyle="-",marker="o",alpha=0.9,linewidth=2,zorder=5)
plt.title("line plot")
plt.legend(loc="best")
plt.xlabel("years",fontdict={"size":19,"color":"#3D2B25"})
plt.ylabel("runs")
# plt.xlim(2008,2010)
plt.grid()
plt.xticks(ticks=data['index'], labels=data['index'],rotation=45)
plt.show()
```

```python
data=pd.read_csv(r"E:\Datasets\batsman_season_record.csv")
data.head()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data['batsman'],data['2015'],color=['green','red','blue','yellow','pink'],bottom=0,hatch="x",log=True)
plt.xlabel("batsman")
plt.ylabel('2015 season')
plt.title("Bar plot")
plt.grid()
# plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(np.arange(data.shape[0])-0.2,data['2015'],width=0.2)
plt.bar(np.arange(data.shape[0]),data['2016'],width=0.2)
plt.bar(np.arange(data.shape[0])+0.2,data['2017'],width=0.2)
plt.xlabel("batsman")
plt.ylabel('2015 season')
plt.xticks(np.arange(data.shape[0]),data['batsman'])
plt.title("Bar plot")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(data['batsman'],data['2015'],label="2015")
plt.bar(data['batsman'],data['2016'],bottom=data['2015'],label='2016')
plt.bar(data['batsman'],data['2017'],bottom=data['2015']+data['2016'],label='2017')
plt.xlabel("batsman")
plt.ylabel('season')
plt.legend()
plt.title("Bar plot")
plt.show()
```
